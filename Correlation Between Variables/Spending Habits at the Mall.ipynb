# Do Larger Annual Incomes contribute to larger spending habits at the mall?

```python
import numpy as np
import pandas as pd
from scipy.stats import linregress
```

```python
mall_customers = pd.read_csv('Mall_Customers.csv')
mall_customers.head()
```

```python
IncomevsSpend =  mall_customers[['CustomerID', 'Annual Income (k$)', 'Spending Score (1-100)']]
IncomevsSpend.head()
```

```python
IncomevsSpend.sort_values('Annual Income (k$)', ascending=False).head()
```

```python
IncomevsSpend[IncomevsSpend['Annual Income (k$)']>=99]
```

```python
IncomevsSpend_crosstab = pd.pivot_table(data=mall_customers, values='Spending Score (1-100)', index='CustomerID', columns='Annual Income (k$)')
IncomevsSpend_crosstab.head()
```

```python
a = [99, 99, 101, 101, 103, 103, 103, 103, 113, 113, 120, 120, 126, 126, 137, 137]
b = [39, 97, 24, 68, 17, 85, 23, 69, 8, 91, 16, 79, 28, 74, 18, 83]
linregress(a, b)
```



# First off, it is important to note that the r-value, or the correlation coefficient is interpreted as an absolute value. Based off the obtained value of -0.051341980526906146, it is clear that there is a very weak correlation between annual income and spending habits at the mall for those with large annual incomes (greater than 99k). his is because the closer the r-score converges to 1, the stronger the correlation between the two variables. Therefore, we can conclude that there is absolutely no corrrelation between annual spending and annual net worth for those with large annual incomes.
